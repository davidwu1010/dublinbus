### load trips

```python
df_trips=pd.read_csv("../initial_data/RT_Trips.csv",index_col=0)
df_trips
```

```python
df_trips.columns=['DayOfService','TripID','LineID','RouteID','Direction','PlannedTime_Dep','PlannedTime_Arr','ActualTime_Dep','ActualTime_Arr']
df_trips
```

```python
df_trips_simple=df_trips[['DayOfService','TripID','LineID','RouteID']]
df_trips_simple
```

### merge two dataset

```python
section=pd.merge(df_leaveTime,df_trips_simple,how='left',on=['DayOfService','TripID'])
section
```

### add previous stops

```python
section_copy=section[['TripID','StopPointID','ActualTime_Arr']]
section_copy.columns=['TripID_2','StopPoint_2','ActualTime_Arr_2']
section_copy
```

```python
section_copy.drop(axis=0, index=0,inplace=True)
section_copy
```

```python
# add new rows
section_copy.loc['new_raw'] = '3'
section_copy
```

```python
# adject the index
section_copy.index=np.arange(0,9267830)
section_copy
```

```python
section_combine = pd.concat([section, section_copy], axis=1,ignore_index=True)
```

```python
col_name=[]
for col in section.columns:
    col_name.append(col)
col_name.append('TripID_2')
col_name.append('StopPoint_2')
col_name.append('ActualTime_Arr_2')
col_name
```

```python
section_combine.columns=col_name
section_combine
```

```python
#drop the last rows
section_combine.drop(axis=0, index=9267829,inplace=True)
section_combine
```

```python
section_combine['trip_q']=section_combine['TripID_2']-section_combine['TripID']
```

```python
section_combine
```

```python
section_combine['trip_q'].value_counts()
```

```python
# if trip_q !=0, we do not need it 
x=section_combine[section_combine['trip_q']==0]
```

```python
x['TravelTime']=x['ActualTime_Arr_2']-x['ActualTime_Dep']
x
```

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
plt.hist(x['TravelTime'], bins=400, facecolor="blue", edgecolor="black")
# x
plt.xlabel("TravelTime")
# y
plt.ylabel("frequency")

plt.show()
```

### there are a lot of unmormal value, drop traveltime which is smaller than 0 first

```python
x=x[x['TravelTime']>0]
```

```python
plt.hist(x['TravelTime'], bins=400, facecolor="blue", edgecolor="black")
# x
plt.xlabel("TravelTime")
# y
plt.ylabel("frequency")

plt.show()
```

```python
#drop traveltime >=3000
x=x[x['TravelTime']<3000]
```

```python
plt.hist(x['TravelTime'], bins=400, facecolor="blue", edgecolor="black")
# x
plt.xlabel("TravelTime")
# y
plt.ylabel("frequency")

plt.show()
```
