```python
def binary_search(item,alist):
    n = len(alist)
    start = 0
    end = n - 1
    while start <= end:
        mid = (start + end) // 2
        if alist[mid] == item:
            return False
        elif item < alist[mid]:
            end = mid - 1
        else:
            start = mid + 1
    return True
```

```python
difference_18_20=[]
for stop in stopPointID_18:
    if binary_search(stop,stopPointID_20):
        difference_18_20.append(stop)
len(difference_18_20)
```
